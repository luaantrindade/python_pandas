## Criando Novas Variaveis 

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
   dados.head(10
             )
```

### Chefe pediu para conseguir o valor bruto do total residencial

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

### O Chefe pediu valor por metro ²

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
# Arredondando valores do metros quadrados para 2 casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(10)
```

### Valor bruto por metro ²

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

### Criar um tipo agregado de moradias

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# O apply serve para aplicar as alterações que vc deseja a cada registro do DataFrame
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

# Excluindo Variaveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux.head(10)
```

```python
 # Sempre apagar variaveis nos dados auxiliares(criando variaveis para armazenar auxiliarmente)
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

```python
# Outra forma de excluir Colunas(variaveis)
dados_aux.pop('Valor Bruto m2')
```

```python
# 3 forma de excluir colunas variaveis
# Pode excluir tanto variaveis(colunas), como variaveis(linhas), e o paramentro para identificar isso é o AXIS=0 é linhas e AXIS=1 é Coluna
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis=1, inplace= True)
```

```python
dados.head(10)
```

```python
dados.to_csv('aluguel_residencial.csv', sep=';', index= False)
```
